```python
# Courbe LOSS (visualisation de l'évolution de l'entraînement)

# Importation des bibliothèques nécessaires
import matplotlib.pyplot as plt
import numpy as np
import os

# Création du dossier pour sauvegarder les graphiques
os.makedirs("../images/training_curves", exist_ok=True)

# Définition des époques (iterations d'entraînement)
epochs = [1, 2, 3, 4, 5]

# Valeurs simulées de la loss sur le dataset d'entraînement
train_loss = [0.9, 0.7, 0.5, 0.35, 0.2]

# Valeurs simulées de la loss sur le dataset de validation
val_loss = [0.95, 0.75, 0.6, 0.4, 0.25]

# Création de la figure
plt.figure()

# Courbe de la loss d'entraînement
plt.plot(epochs, train_loss, label="Train Loss")

# Courbe de la loss de validation
plt.plot(epochs, val_loss, label="Val Loss")

# Titre du graphique
plt.title("Training Loss Curve")

# Affichage de la légende
plt.legend()

# Sauvegarde du graphique dans un fichier image
plt.savefig("../images/training_curves/loss_curve.png")

# Affichage du graphique
plt.show()
```

```python
# Courbe ACCURACY (visualisation de l'évolution de la précision)

# Importation des bibliothèques nécessaires
import matplotlib.pyplot as plt
import os

# Création du dossier pour sauvegarder les graphiques
os.makedirs("../images/training_curves", exist_ok=True)

# Valeurs simulées de l'accuracy sur l'entraînement
train_acc = [0.60, 0.70, 0.78, 0.85, 0.92]

# Valeurs simulées de l'accuracy sur la validation
val_acc = [0.58, 0.68, 0.75, 0.82, 0.88]

# Création de la figure
plt.figure()

# Courbe accuracy entraînement
plt.plot(epochs, train_acc, label="Train Accuracy")

# Courbe accuracy validation
plt.plot(epochs, val_acc, label="Val Accuracy")

# Titre du graphique
plt.title("Accuracy Curve")

# Légende
plt.legend()

# Sauvegarde de l'image
plt.savefig("../images/training_curves/accuracy_curve.png")

# Affichage
plt.show()
```

```python
# Visualisation d'une carte d'attention (attention map)

# Importation des bibliothèques nécessaires
import matplotlib.pyplot as plt
import numpy as np
import os

# Création du dossier de sauvegarde si non existant
os.makedirs("../images/attention_maps", exist_ok=True)

# Génération d'une matrice d'attention aléatoire (8x8)
attention = np.random.rand(8, 8)

# Création de la figure
plt.figure()

# Affichage de la matrice d'attention
plt.imshow(attention, cmap="viridis")

# Titre du graphique
plt.title("Attention Map")

# Barre de couleur
plt.colorbar()

# Sauvegarde de l'image
plt.savefig("../images/attention_maps/attention_map.png")

# Affichage
plt.show()
```
