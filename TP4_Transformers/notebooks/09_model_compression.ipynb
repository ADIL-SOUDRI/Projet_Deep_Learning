```python
# Importation des bibliothèques nécessaires
import torch
import time
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
# Définition d'un dictionnaire de modèles NLP à tester
# Chaque clé représente un nom logique, et la valeur le modèle Hugging Face associé
models = {
    "bert": "bert-base-uncased",
    "distilbert": "distilbert-base-uncased"
}
```

```python
# Liste de phrases pour tester un modèle NLP en mode inference
# Chaque phrase sera utilisée pour vérifier la prédiction du modèle
texts = [
    "I love deep learning",
    "Transformers are powerful models",
    "NLP is fascinating",
    "I hate bugs in software",
    "Machine learning is amazing"
]
```

```python
# Fonction de test pour comparer les performances de différents modèles NLP
def test_model(model_name):

    # Chargement du tokenizer associé au modèle
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Chargement du modèle de classification de séquences
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=2
    )

    # Début du chronométrage
    start = time.time()

    # Désactivation du calcul des gradients pour accélérer l'inférence
    with torch.no_grad():

        # Parcours de chaque texte à tester
        for text in texts:

            # Tokenisation du texte en entrée
            inputs = tokenizer(
                text,
                return_tensors="pt"
            )

            # Passage dans le modèle pour obtenir les prédictions
            outputs = model(**inputs)

    # Fin du chronométrage
    end = time.time()

    # Retour du temps total d'exécution (benchmark)
    return end - start
```

```python
# Dictionnaire pour stocker les temps d'exécution de chaque modèle
results = {}

# Boucle sur les modèles définis dans le dictionnaire "models"
for name, model_name in models.items():

    # Mesure du temps d'exécution du modèle
    duration = test_model(model_name)

    # Stockage du résultat dans le dictionnaire
    results[name] = duration

    # Affichage du temps pour chaque modèle
    print(f"{name} time: {duration:.4f} sec")
```

```python
# Affichage de l'analyse des performances des modèles testés
print("\n=== ANALYSIS ===")

# Affichage du temps d'exécution de BERT
print("BERT       :", results["bert"])

# Affichage du temps d'exécution de DistilBERT
print("DistilBERT :", results["distilbert"])

# Calcul du gain de performance (speedup)
# > 1 signifie que DistilBERT est plus rapide que BERT
speedup = results["bert"] / results["distilbert"]

# Affichage du facteur d'accélération
print(f"Speedup : {speedup:.2f}x")
```
