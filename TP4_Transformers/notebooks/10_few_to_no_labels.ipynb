```python
# Importation de la fonction pipeline de Hugging Face
from transformers import pipeline

# Création d'un pipeline de classification zero-shot
# Ce modèle peut classer du texte sans entraînement spécifique sur tes labels
classifier = pipeline(
    "zero-shot-classification",
    model="typeform/distilbert-base-uncased-mnli"
)

# Confirmation du chargement du modèle
print("Model loaded successfully")
```

```python
# Liste des labels candidats pour la classification zero-shot
# Le modèle va choisir la classe la plus probable sans entraînement spécifique
candidate_labels = [
    "positive",
    "negative"
]
```

```python
# Liste de textes à classer avec un modèle de classification zero-shot
texts = [
    "I love deep learning",
    "Transformers are powerful models",
    "I hate bugs in software",
    "This code is terrible",
    "Machine learning is amazing"
]
```

```python
# Boucle sur chaque texte pour appliquer la classification zero-shot
for text in texts:

    # Application du modèle zero-shot classification
    result = classifier(
        text,
        candidate_labels
    )

    # Affichage du texte analysé
    print("\nText :", text)

    # Affichage du label prédit (le plus probable)
    print("Prediction :", result["labels"][0])

    # Affichage du score de confiance associé à la prédiction
    print("Score :", round(result["scores"][0], 4))
```

```python
# Importation de la fonction pipeline de Hugging Face
from transformers import pipeline

# Création d'un pipeline de génération de texte
# Utilisation du modèle distilgpt2 (version légère de GPT-2)
generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
# Prompt de génération de texte pour classification via text-generation
# On donne des exemples (few-shot learning) pour guider le modèle
prompt = """
Text: I love AI
Label: positive

Text: I hate bugs
Label: negative

Text: Machine learning is amazing
Label:
"""
```

```python
# Génération de texte avec le modèle GPT-2 (distilgpt2)
result = generator(
    prompt,
    max_length=60,   # longueur maximale du texte généré
    do_sample=False  # génération déterministe (sans random)
)

# Affichage du texte généré complet
print(result[0]["generated_text"])
```
