```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_name = "t5-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

print("T5 model loaded successfully")
```

```python
prompt = "summarize: Deep learning is a subset of machine learning that uses neural networks."

inputs = tokenizer(
    prompt,
    return_tensors="pt",
    truncation=True
)

with torch.no_grad():
    output = model.generate(
        **inputs,
        max_length=60
    )

generated_text = tokenizer.decode(
    output[0],
    skip_special_tokens=True
)

print(generated_text)
```

```python
prompts = [
    "summarize: Artificial intelligence is transforming many industries.",
    "summarize: Transformers are powerful neural network architectures.",
    "summarize: Deep learning helps computers learn from data."
]

for p in prompts:

    inputs = tokenizer(
        p,
        return_tensors="pt",
        truncation=True
    )

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_length=40
        )

    result = tokenizer.decode(
        output[0],
        skip_special_tokens=True
    )

    print("\nPrompt:")
    print(p)

    print("Generated:")
    print(result)
```

```python
test_texts = [
    "summarize: Natural Language Processing allows computers to understand human language.",
    "summarize: Machine learning is widely used in recommendation systems.",
    "summarize: Large language models are changing software development."
]

for text in test_texts:

    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True
    )

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_length=50
        )

    print("\nInput:")
    print(text)

    print("Summary:")
    print(tokenizer.decode(
        output[0],
        skip_special_tokens=True
    ))
```
