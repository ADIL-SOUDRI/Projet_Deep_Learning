```python
# Importation de la fonction pipeline de Hugging Face
from transformers import pipeline

# Création d'un pipeline de Question Answering (QA)
# Modèle pré-entraîné sur SQuAD pour répondre à des questions
qa = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

# Contexte dans lequel le modèle va chercher la réponse
context = """
Transformers are deep learning models introduced by Google researchers in 2017.
They rely on the self-attention mechanism to capture relationships between words.
Today they are widely used in Natural Language Processing tasks such as
text classification, machine translation, question answering,
text generation and summarization.
"""

# Question posée au modèle
question = "When were Transformers introduced?"

# Exécution du modèle QA
result = qa(
    question=question,
    context=context
)

# Affichage de la réponse extraite du contexte
print("Answer:", result["answer"])

# Affichage du score de confiance de la réponse
print("Score:", result["score"])
```

```python
# Liste de questions pour tester le modèle de question-answering
questions = [
    "When were Transformers introduced?",
    "What mechanism do Transformers use?",
    "Which NLP tasks use Transformers?",
    "Who introduced Transformers?"
]

# Boucle sur chaque question
for q in questions:

    # Passage de la question et du contexte au modèle QA
    result = qa(
        question=q,
        context=context
    )

    # Affichage de la question posée
    print("\nQuestion:", q)

    # Affichage de la réponse extraite par le modèle
    print("Answer:", result["answer"])

    # Affichage du score de confiance de la réponse
    print("Score:", result["score"])
```
