```python
from transformers import pipeline

qa = pipeline("question-answering")

context = """
Transformers are deep learning models based on the attention mechanism.
They are widely used in NLP tasks such as question answering, translation,
and text summarization. The architecture was introduced in 2017.
"""

question = "When were Transformers introduced?"

result = qa(question=question, context=context)

print("Answer:", result["answer"])
print("Score:", result["score"])
```

```python
questions = [
    "What are Transformers based on?",
    "What tasks use Transformers?",
    "When were they introduced?"
]

for q in questions:
    result = qa(question=q, context=context)

    print("\nQuestion:", q)
    print("Answer:", result["answer"])
```
