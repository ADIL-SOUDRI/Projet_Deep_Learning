```python
import torch
from transformers import pipeline

print("Torch OK:", torch.__version__)
print("Transformers OK")

summarizer = pipeline(
    "summarization",
    model="t5-small"
)

print("Pipeline OK")

text = """
Artificial intelligence and deep learning have transformed many industries.
Transformers are a powerful architecture used in NLP tasks such as translation,
summarization and text generation.
"""

summary = summarizer(
    text,
    max_length=40,
    min_length=15,
    do_sample=False
)

print(summary[0]["summary_text"])
# from transformers import pipeline
```

```python
text = """
Artificial intelligence and deep learning have transformed many industries.
Transformers are a powerful architecture used in NLP tasks such as translation,
summarization and text generation. They rely on self-attention mechanisms to
capture relationships between words in a sentence.
"""
```

```python
summary = summarizer(
    text,
    max_length=40,
    min_length=15,
    do_sample=False
)

print(summary[0]["summary_text"])
```

```python
texts = [
    """
    Machine learning is a subset of artificial intelligence that allows systems
    to learn from data and improve over time without being explicitly programmed.
    """,
    """
    Transformers use attention mechanisms to process sequences of data in parallel
    and have become state-of-the-art in NLP tasks.
    """
]

for t in texts:
    result = summarizer(t, max_length=40, min_length=10)
    print("\nOriginal:", t)
    print("Summary:", result[0]["summary_text"])
```
