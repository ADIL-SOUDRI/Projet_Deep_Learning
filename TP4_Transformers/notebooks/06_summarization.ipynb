```python
# Importation de PyTorch pour vérifier l'installation et gérer les tenseurs
import torch

# Importation de la fonction pipeline de Hugging Face
from transformers import pipeline

# Vérification de la version de PyTorch installée
print("Torch OK:", torch.__version__)

# Confirmation que Transformers est bien importé
print("Transformers OK")

# Création d'un pipeline de summarization basé sur le modèle T5-small
summarizer = pipeline(
    "summarization",
    model="t5-small"
)

# Vérification que le pipeline est prêt à être utilisé
print("Pipeline OK")

# Texte d'entrée à résumer
text = """
Artificial intelligence and deep learning have transformed many industries.
Transformers are a powerful architecture used in NLP tasks such as translation,
summarization and text generation.
"""

# Génération du résumé avec le pipeline
summary = summarizer(
    text,
    max_length=40,   # longueur maximale du résumé
    min_length=15,   # longueur minimale du résumé
    do_sample=False  # génération déterministe (sans random)
)

# Affichage du résumé généré
print(summary[0]["summary_text"])

# from transformers import pipeline
```

```python
# Texte d'entrée à résumer
# Ce texte décrit l'intelligence artificielle, les Transformers et leurs applications en NLP
text = """
Artificial intelligence and deep learning have transformed many industries.
Transformers are a powerful architecture used in NLP tasks such as translation,
summarization and text generation. They rely on self-attention mechanisms to
capture relationships between words in a sentence.
"""
```

```python
# Génération du résumé du texte avec le pipeline NLP
summary = summarizer(
    text,
    max_length=40,   # longueur maximale du résumé généré
    min_length=15,   # longueur minimale du résumé
    do_sample=False  # génération déterministe (sans random)
)

# Affichage du résumé généré
print(summary[0]["summary_text"])
```

```python
# Liste de textes à résumer
texts = [
    """
    Machine learning is a subset of artificial intelligence that allows systems
    to learn from data and improve over time without being explicitly programmed.
    """,
    """
    Transformers use attention mechanisms to process sequences of data in parallel
    and have become state-of-the-art in NLP tasks.
    """
]

# Boucle sur chaque texte pour générer un résumé
for t in texts:

    # Génération du résumé avec le pipeline summarization
    result = summarizer(
        t,
        max_length=40,   # longueur maximale du résumé
        min_length=10    # longueur minimale du résumé
    )

    # Affichage du texte original
    print("\nOriginal:", t)

    # Affichage du résumé généré
    print("Summary:", result[0]["summary_text"])
```
